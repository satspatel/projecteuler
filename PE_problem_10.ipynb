```
s = 5
for i in range(1999999, 4, -2): # start from 5 and increment up in 2s to 2 million
    half = int(round(i/3,0)) # returns a third of the value of i
    if half % 2 == 0:        # a prime won't divide by an even number so work out if it is
        half +=1             # if it is even then add 1 to make it odd
    for d in range(1, half+2, 2): #starting from 1 work up to half + 2
        if i % d == 0:            # if d is a multiple of i
            if d == 1:            # skip 1
                pass
            else:                 # if d is a multiple if i then go to the next i
                break
        elif d == half:           # if 'half' has been reached then it is a prime
            s+=i                  # add the prime to the sum
            break
    if i % 1001 == 0:
        print(str(i) + ' done. Sum of primes is:' + str(s))
print(i)
print(s)
```
